```python
import spacy

# Load Norwegian language model
nlp = spacy.load('nb_core_news_lg') 
# nlp.max_length = 1005000 # 1004420

# Define function to lemmatize Norwegian text
def lemmatize_dont_take_pos_into_account(list_of_texts):
    # Parse the text with the loaded language model
    docs = list(nlp.pipe(list_of_texts))
    # Extract the lemmatized form of each token in the text
    # https://stackoverflow.com/questions/57747613/punctuation-stopwords-and-lemmatization-with-spacy
    lemmas = [token.lemma_ for doc in docs for token in doc if not token.is_punct and not token.like_num]
    print("Lemmas:", lemmas[:100])
    # Join the lemmatized words back into a string
    lemmatized_text = ' '.join(lemmas)
    print("Lemmatized_text:", lemmatized_text[:1000])
    
    return lemmatized_text

def lemmatize(list_of_texts):
  docs = list(nlp.pipe(list_of_texts))
  lemmas_list = [(token.lemma_,token.pos_) for doc in docs for token in doc if not token.is_punct and not token.like_num]
  print("lemmas_list:", lemmas_list[:100])
  return lemmas_list
  
# Get part of speech for each word in the dataset
def get_pos(list_of_texts):
  docs = list(nlp.pipe(list_of_texts))
  pos = [(token.text, token.pos_) for doc in docs for token in doc if not token.is_punct and not token.like_num]
  print("PoS:", pos[:100])
  return pos # flattened list
```
